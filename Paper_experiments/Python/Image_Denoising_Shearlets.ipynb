```python
import shearlab
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import time
from scipy import ndimage as img
from scipy import io as sio
```

```python
from shearlab import load_image, imageplot
```

### Data loading

```python
# load data
data = load_image("../data_samples/boat.png", 512)
data = data.astype(float)*255
```

```python
plt.figure(figsize=[6,6])
plt.axis('off')
plt.imshow(data, cmap='gray')
```

```python
sigma = 30
scales = 4
thresholdingFactor = 3
X = data
```

```python
sizeX = data.shape[0]
sizeY = data.shape[1]
rows = sizeX
cols = sizeY
X = data;
```

```python
# add noise
X_noisy = X + sigma*np.random.randn(X.shape[0], X.shape[1])
```

### Decomposition

```python
nScales = 4;
shearLevels = [float(shearlab.ceil(i/2)) for i in range(1,nScales+1)]
scalingFilter = 'Shearlab.filt_gen("scaling_shearlet")'
directionalFilter = 'Shearlab.filt_gen("directional_shearlet")'
waveletFilter = 'Shearlab.mirror(scalingFilter)'
scalingFilter2 = 'scalingFilter'
full = 0;
```

```python
# Generating the shearlet system
shearletSystem = shearlab.getshearletsystem2D(rows,cols,nScales,shearLevels,full,directionalFilter,scalingFilter);
```

```python
# Computing coefficients
coeffs = shearlab.sheardec2D(X_noisy,shearletSystem)
```

### Denoising by hard thresholding

```python
oldCoeffs = coeffs.copy()
weights = np.ones(coeffs.shape)
```

```python
for j in range(len(shearletSystem.RMS)):
    weights[:,:,j] = shearletSystem.RMS[j]*np.ones((X.shape[0], X.shape[1]))
```

```python
coeffs = np.real(coeffs)
zero_indices = np.abs(coeffs) / (thresholdingFactor * weights * sigma) < 1
coeffs[zero_indices] = 0
```

### Reconstruction

```python
# reconstruction
Xrec = shearlab.shearrec2D(coeffs, shearletSystem)
```

```python
from skimage.measure import compare_psnr as psnr
```

```python
PSNR = psnr(X/X.max(), Xrec/Xrec.max())
print("{0:.2f}".format(PSNR))
```
