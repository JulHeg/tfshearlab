```python
import shearlab
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import time
from scipy import ndimage as img
from scipy import io as sio
```

```python
from shearlab import load_image, imageplot
```

### Data loading

```python
# load data
data = load_image("../data_samples/boat.png", 512)
data = data.astype(float)*255
```

```python
plt.figure(figsize=[6,6])
plt.axis('off')
plt.imshow(data, cmap='gray')
```

```python
sigma = 30
scales = 4
thresholdingFactor = 3
X = data
```

```python
sizeX = data.shape[0]
sizeY = data.shape[1]
rows = sizeX
cols = sizeY
X = data;
```

```python
# add noise
X_noisy = X + sigma*np.random.randn(X.shape[0], X.shape[1])
```

```python
plt.figure(figsize=[6,6])
plt.axis('off')
plt.imshow(X_noisy, cmap='gray')
```

### Denoising via BM3D

```python
from bm3d import BM3D_1st_step, BM3D_2nd_step
```

```python
Basic_img = BM3D_1st_step(X_noisy)
```

```python
plt.figure(figsize=[6,6])
plt.axis('off')
plt.imshow(Basic_img, cmap='gray')
plt.savefig('../results/denoising/denoised1_bm3d.png')
```

```python
Final_img = BM3D_2nd_step(Basic_img, X_noisy)
```

```python
plt.figure(figsize=[6,6])
plt.axis('off')
plt.imshow(Final_img, cmap='gray')
plt.savefig('../results/denoising/denoised2_bm3d.png')
```

```python
from skimage.measure import compare_psnr as psnr
```

```python
PSNR = psnr(X/X.max(), Basic_img/Basic_img.max())
print("{0:.2f}".format(PSNR))
```

```python
PSNR = psnr(X/X.max(), Final_img/Final_img.max())
print("{0:.2f}".format(PSNR))
```
